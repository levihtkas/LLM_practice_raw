```python
!pip install beautifulsoup4
!pip install requests

from bs4 import BeautifulSoup
import requests

def fetch_website_contents(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.content,"html.parser")
  content_output = soup.body.get_text(separator="\n", strip=True)
  return content_output
 
```

```python
contents = fetch_website_contents("https://edwarddonner.com")
```

```python
def messages_format(content):
  messages = [
  {"role":"system","content":"You a assistant analysis the content of website and share related infomartion that relates to AI , machine learning"},
  {"role":"user","content":content}
]
  return messages
```

```python
def summarizer(url):
  website_contents =fetch_website_contents(url)
  response = openai.chat.completions.create(
    model = "gpt-5-nano",
    messages = messages_format(website_contents)
  ) 
  return response.choices[0].message.content
```

```python
summarizer("https://edwarddonner.com")
```
