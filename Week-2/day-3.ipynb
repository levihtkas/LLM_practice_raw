```python
# imports
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
# Load environment variables in a file called .env
# Print the key prefixes to help with any debugging

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
```

```python
# Initialize

openai = OpenAI()
MODEL = 'gpt-4.1-mini'
```

```python
# Again, I'll be in scientist-mode and change this global during the lab

system_message = "You are a helpful assistant"
```

```python
def chat(message, history):
    return f"You said {message} and the history is {history} but I still say bananas"
gr.ChatInterface(fn=chat).launch()
```

```python
def chat(message,history):
  print(history)
  history = [{"role":h['role'],"content":h['content'][0]['text']} for h in history]
  messages = [{"role":'system',"content":system_message}] + history + [{'role':'user','content':message}]
  stream = openai.chat.completions.create(model=MODEL,messages=messages,stream=True)
  response =""
  for chunk in stream:
    delta=chunk.choices[0].delta
    if delta.content:
      response+=delta.content
      yield response

```

```python
# for our chat bot we need to get the chat updated


gr.ChatInterface(fn=chat).launch()
```
